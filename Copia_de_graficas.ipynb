# Gráfica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return x - 3**(-x)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
#plt.axhline(0, color="black")
#plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-5, 5)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```

Ejercicio 1: Graficar las funciones $x$ y $3^{-x}$ en una sola figura y comentar cómo su resta da origen a la gráfica de $f(x)$.

Ejercicio 2: Agregar título, color, retícula, rótulos a los ejes y aumentar el tamaño de la gráfica.

```python
def g(x):
  return x

def h(x):
  return 3**(-x)

plt.figure(figsize=(10,5))
plt.plot(x,f(x),color="black")
plt.plot(x,g(x))
plt.plot(x,h(x))
plt.xlim(-10,10)
plt.ylim(-5,5)
plt.title("Prueba $f(x)=x-3^{-x}$")
plt.xlabel("Eje X")
plt.ylabel("Eje y")
plt.grid()
#plt.savefig("output.png")
plt.show()
```

Aquí se trató de mezclar las funciones $g(x)$ y $h(x)$ tal que $f(x)=g(x)+h(x)$
