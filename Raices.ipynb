#Ejercicio 1
Encuentre el Polinomio de Taylor de tercer grado para $f(x)=\sqrt{x+1}$ respecto a $x_0=0$

Aproxime $\sqrt{0.5}$, $\sqrt{0.75}$, $\sqrt{0.125}$ y $\sqrt{1.5}$; y encuentre los valores reales.


Primero vemos que $f(x)$ es continua en el rango $(-1,∞)$.
Ahora calculamos hasta la cuarta derivada de $f(x)$:

$f(x)'=\frac{1}{2\sqrt{x+1}}$

$f(x)''=-\frac{1}{4(x+1)^{3/2}}$

$f(x)'''=\frac{3}{8(x+1)^{5/2}}$

$f(x)^{IV}=-\frac{15}{16(x+1)^{7/2}}$

Dado que el Polinomio de Taylor se define como: $P_n(x)=∑^n_{k=0}\frac{f^k(x_0)}{k!}(x-x_0)^k$

$P_0(x)=\sqrt{0+1}\frac{1}{0!}(x-0)^0=1$

$P_1(x)=P_0(x)+\frac{1}{2\sqrt{0+1}}\frac{1}{1!}(x-0)^1=1+\frac{1}{2}x$

$P_2(x)=P_1(x)+(-\frac{1}{4(0+1)^{3/2}})\frac{1}{2!}(x-0)^2=1+\frac{1}{2}x-\frac{1}{8}x^2$

Tenemos que para $P_3(x)$:

$P_3(x)=P_2(x)+ \frac{3}{8(0+1)^{5/2}}\frac{1}{3!}(x-0)^3=1+\frac{1}{2}x-\frac{1}{8}x^2+\frac{1}{16}x^3$


Y por ende el residuo es:

$R_3(x)=\frac{f^{(3+1)}(\xi(x))}{(3+1)!}(x-0)^{3+1}=\frac{-\frac{15}{16(\xi(x)+1)^{7/2}}x^4}{24}$



```python


import matplotlib.pyplot as plt
import numpy as np


def P0(x):
  return 1

def P1(x):
  return 1+(x/2)

def P2(x):
  return 1+(x/2)-(x**2/8)

def P3(x):
  return 1+(x/2)-(x**2/8)+(x**3/16)

def raiz(x):
  return x**(1/2)

def difpr(G,H):
  return G-H

x=np.linspace(-1, 1,100)
```

###a)$\sqrt{0.5}$; Para resolver, buscamos $x$ tal que $x+1=0.5$, en este caso $x=-0.5$


```python
print("El Polinomio grado 3 es:", P3(-0.5))
print("La raíz es:", raiz(0.5))
print("Y el error real es de:", difpr(P3(-0.5),raiz(0.5)))
```

###b)$\sqrt{0.75}$; Para resolver, buscamos $x$ tal que $x+1=0.75$, en este caso $x=-0.25$


```python
print("El Polinomio grado 3 es:", P3(-0.25))
print("La raíz es:", raiz(0.75))
print("Y el error real es de:", difpr(P3(-0.25),raiz(0.75)))
```

###c)$\sqrt{0.125}$; Para resolver, buscamos $x$ tal que $x+1=0.125$, en este caso $x=-0.875$


```python
print("El Polinomio grado 3 es:", P3(-0.875))
print("La raíz es:", raiz(0.125))
print("Y el error real es de:", difpr(P3(-0.875),raiz(0.125)))
```

###d)$\sqrt{1.5}$; Para resolver, buscamos $x$ tal que $x+1=1.5$, en este caso $x=0.5$

```python
print("El Polinomio grado 3 es:", P3(0.5))
print("La raíz es:", raiz(1.5))
print("Y el error real es de:", difpr(P3(0.5),raiz(1.5)))
```
